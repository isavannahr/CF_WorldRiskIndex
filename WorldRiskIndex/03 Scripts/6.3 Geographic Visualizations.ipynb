# 04 Chloropleth

```python
data_to_plot = df[['Region','WRI']]
data_to_plot.head()
```

```python
# Setup a folium map at a high-level zoom
map = folium.Map(location = [100, 0], zoom_start = 1.5)

# Choropleth maps bind Pandas Data Frames and json geometries.This allows us to quickly visualize data combinations
folium.Choropleth(
    geo_data = country_geo, 
    data = data_to_plot,
    columns = ['Region', 'WRI'],
    key_on = 'feature.properties.name', # this part is very important - check your json file to see where the KEY is located
    fill_color = 'YlOrBr', fill_opacity=1, line_opacity=0.1,     nan_fill_color="lightgrey",
    nan_fill_opacity=0.4,
    legend_name = "WRI").add_to(map)
folium.LayerControl().add_to(map)

map
```

Countries  with higher WRI appear to be along the equator. My hypothesis is that this is due to higher exposure rates due to natural disaster such as heat, weather patterns, disease due to instects and other wildlife along with other factors.
